### Step 4. How many products cost more than $10.00?

```python
chipo["item_price"]=chipo["item_price"].str.replace("$","")
chipo["item_price"]=pd.to_numeric(chipo["item_price"])
```

```python
chipo_price_greater_10=(chipo["item_price"]>10).sum()
print(chipo_price_greater_10)
```

### Step 6. Sort by the name of the item

```python
sorted_value=chipo.sort_values(by=["item_name"])
sorted_value.head(20)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
quantity=chipo["quantity"].loc[chipo["item_price"]==chipo["item_price"].max()]
quantity
```

### Step 8. How many times were a Veggie Salad Bowl ordered?

```python
times=chipo[chipo["item_name"]=="Veggie Salad Bowl" ]["quantity"].count()
times
```

### Step 9. How many times people orderd more than one Canned Soda?

```python
order=chipo[(chipo["quantity"]>1) &(chipo["item_name"]=="Canned Soda")]
order["quantity"].count()
```
