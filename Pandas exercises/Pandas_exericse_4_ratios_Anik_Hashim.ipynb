### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', delimiter ='|')
users.set_index('user_id',inplace=True)
users
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
dd = (users.groupby(['occupation'])['gender'].value_counts(99)).sort_values(ascending = False).xs('M', level=1).to_frame()
dd.columns = ['gender_ratio']
dd
#dd = dd[dd.gender != 'F']

# rak = (users.groupby(['occupation', 'gender'])['gender'].count())
# esh = users.groupby(['occupation']).gender.count()
# maha = rak/esh 
# to = maha.sort_values(ascending=False).xs("M",level=1)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
### aggregate function can replace both these lines.
# users.groupby('occupation').min()
# users.groupby('occupation').max()

users.groupby('occupation')[('age')].agg([ 'min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
u = users.groupby(['occupation','gender'])['age'].mean()
u.to_frame()
```

### Step 8.  For each occupation present the percentage of women and men

```python
uss=users.groupby(['occupation'])['gender'].value_counts('M')
uss=uss.to_frame()*100
uss
```
