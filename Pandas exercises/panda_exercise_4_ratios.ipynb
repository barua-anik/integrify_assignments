### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
user=pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user",sep="|")
user.head(10)
```

### Step 4. Discover what is the mean age per occupation

```python
user.groupby("occupation")["age"].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python

i = user.groupby(['occupation' ,'gender']).gender.count() 
total_M_F = user.groupby('occupation').gender.count()
(i / total_M_F).sort_values(ascending=False).xs('M', level=1)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
user.groupby("occupation").age.agg(["min","max"])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
user.groupby(['occupation','gender']).age.mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
#user.groupby(["occupation","gender"]).occupation.count()

k = user.groupby(['occupation' ,'gender']).occupation.count() 
l = user.groupby('gender').gender.count()
k.head()
```
